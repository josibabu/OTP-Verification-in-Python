```python
import random  # Importing the random module for generating OTP
import smtplib  # Importing smtplib for sending emails
import ssl  # Importing SSL for secure connection
import tkinter as tk  # Importing tkinter for GUI
from tkinter import simpledialog, messagebox  # Importing specific modules from tkinter

MAX_ATTEMPTS = 5  # Maximum number of attempts allowed for OTP verification

# Function to send OTP to the given email address
def send_otp(email):
    port = 587  # Port for STARTTLS
    smtp_server = "smtp.gmail.com"  # SMTP server for Gmail
    sender_email = "projectdata333@gmail.com"  # Sender's email address
    password = "rjvu zhgu uous hhvk"  # Sender's email password (for authentication)

    # Generating a random OTP of 6 digits
    otp = ''.join([str(random.randint(0, 9)) for i in range(6)])

    # Constructing the email message with OTP
    message = f"""\
    Subject: OTP Verification

    Your OTP is {otp}.

    This OTP is valid for 5 minutes. Do not share it with anyone.
    """

    # Creating SSL context for secure connection
    context = ssl.create_default_context()
    with smtplib.SMTP(smtp_server, port) as server:
        # Initiating a secure connection
        server.starttls(context=context)
        # Logging in to the email server
        server.login(sender_email, password)
        # Sending the email with OTP
        server.sendmail(sender_email, email, message)

    return otp  # Returning the generated OTP

# Function to verify the entered OTP
def verify_otp(email, sent_otp):
    attempts = 0  # Initializing the number of attempts
    while attempts < MAX_ATTEMPTS:
        # Asking the user to enter the OTP
        entered_otp = simpledialog.askstring("OTP Verification", "Enter OTP sent to your email:")
        # Checking if the entered OTP matches the sent OTP
        if entered_otp == sent_otp:
            # Showing a success message if OTP matches
            messagebox.showinfo("OTP Verification", "OTP verification successful.")
            return
        else:
            # Showing an error message if OTP does not match
            messagebox.showerror("OTP Verification", "Invalid OTP. Please try again.")
            attempts += 1  # Incrementing the attempt count

    # Showing an error message if maximum attempts reached
    messagebox.showerror("OTP Verification", f"Maximum OTP attempts ({MAX_ATTEMPTS}) reached. Exiting.")

# Function to initiate OTP sending process through GUI
def send_otp_gui():
    # Asking the user to enter their email address
    email = simpledialog.askstring("Email Address", "Enter your email address:")
    if email:
        # Sending OTP to the provided email address
        sent_otp = send_otp(email)
        # Verifying the entered OTP
        verify_otp(email, sent_otp)

# Main function to run the GUI application
if __name__ == "__main__":
    root = tk.Tk()  # Creating the main window
    root.title("Email OTP Verification")  # Setting the title of the window
    send_button = tk.Button(root, text="Send OTP", command=send_otp_gui)  # Creating a button to trigger OTP sending process
    send_button.pack(pady=10)  # Placing the button in the window
    root.mainloop()  # Running the main event loop for the GUI
```
